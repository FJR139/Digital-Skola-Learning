# Package

```python
import os

# create package directory / folder
os.makedirs('mypackage', exist_ok=True)

# Create module1.py
with open('mypackage/module1.py', 'w') as f:
  f.write("""
def add(a, b):
    return a + b
    """)

# Create module2.py
with open ('mypackage/module2.py', 'w') as f:
  f.write("""
def subtract(a, b):
    return a - b
    """)
# Create an empy_init_.py file to mark the directory as a package
with open('mypackage/__init__.py', 'w') as f:
  f.write("")
```

```python
from mypackage import module1, module2

# Use the Functions from the package modules
print(module1.add(5, 3))
print(module2.subtract(5, 3))
```
